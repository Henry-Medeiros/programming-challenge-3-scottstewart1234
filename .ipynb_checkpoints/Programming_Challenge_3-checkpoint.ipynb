```python
# The image below corresponds to a snapshot of single cell organisms observed with a microscope
import cv2
import matplotlib.pyplot as plt
import numpy as np
!wget "https://raw.githubusercontent.com/Henry-Medeiros/images/master/organisms.jpg"
img = cv2.imread('organisms.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY, 0) 

# visualization only
plt.imshow(img, 'gray')
plt.title('Original image')
plt.show()
```

```python
# Binarize the image to separate the organisms from the background
# Hint 1: You may need to remove some high-frequency noise from the image 
# before thresholding
# Hint 2: The background illumination in the image varies significantly, so a 
# single treshold for the entire will not work

#This is a general funciton i made to create a mask to remove high frequency noise from a fourier 
def idealLowPassFilterMask(radius, image):
    xv,yv = np.meshgrid(range(image.shape[1]),range(image.shape[0]))
    euclidian = np.linalg.norm([xv-image.shape[1]/2,yv-image.shape[0]/2],axis=0)
    print(np.amin(euclidian))
    print(np.amax(euclidian))

    ilpf = euclidian < radius

    return ilpf

lowpass = np.fft.ifft2(np.fft.ifftshift(np.fft.fftshift(np.fft.fft2(img)) * idealLowPassFilterMask(100,img)))
lowpass = np.absolute(lowpass) #the image is turned into a fourier, shifted, then multiplied by the low pass mask, then the reverse is done.
#plt.imshow(lowpass,'gray') #uncomment to see the filtered image
cv2.imwrite("./temp.PNG",lowpass)
lowpass = cv2.imread("./temp.PNG")
lowpass = cv2.cvtColor(lowpass, cv2.COLOR_BGR2GRAY ) #save and reload the image just to make cv2 recognize it better. then reconvert to bgr

#use adaptive thresholding so the background is properly filtered despite the changes in the illumination
thresholded = cv2.adaptiveThreshold(lowpass,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV,27,2)
#show this result
plt.imshow(thresholded,'gray')
```

```python
# Locate and plot the organisms present in the image
# Hint 3: You will need to apply a morphological filter to your binarized image 
# to remove spurious pixels left over by the binarization step
# morphological filter to close the objects


#create a kernal to erode the image to leave only the organisms
kernel = np.ones((7,7),np.uint8)
#erode the image to leave only the organisms
erosion = cv2.erode(thresholded,kernel,iterations = 1)


#plot the organisms
plt.imshow(erosion,'gray')
```

```python
# Count the the number of organisms present in the image and compute their 
# average area in pixels

connectivity = 8  #use 8 connectivity to locate insects

#run CCA and get the stats
num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(erosion , connectivity , cv2.CV_32S)

#print the results, subtract 1 because the 0th element of the CCA is the background.
print("The number of organisms preset in the image is :"+str(num_labels-1))

#count the number of pixels in each, then divide by the number of labels and print the results
total = 0
for i in range(1,num_labels):
    total+=stats[i,cv2.CC_STAT_AREA]
    #print(stats[i,cv2.CC_STAT_AREA])
print("and there are an average of "+str(total/(num_labels-1))+" pixels per organism")
```
